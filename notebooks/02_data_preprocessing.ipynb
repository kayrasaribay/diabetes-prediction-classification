# 02 - Data Preprocessing

The goal of this notebook is to clean the dataset and prepare it for machine learning models. The preprocessing steps include removing duplicate rows, handling rare categorical values, separating features and target variable, identifying numerical and categorical columns, and preparing a preprocessing strategy for model training.

## 1. Notebook Objective

The goal of this notebook is to clean the raw diabetes prediction dataset and prepare it for the model training stage.

In the previous notebook, the dataset was explored and some data quality issues were identified, such as fully duplicated rows and a rare category in the `gender` column. In this notebook, these issues will be handled.

The main preprocessing steps in this notebook are:

- Loading the raw dataset
- Removing duplicate rows
- Handling rare categorical values
- Checking missing values
- Encoding the binary `gender` feature
- Identifying numerical and categorical features
- Saving the cleaned dataset for model training

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
```

## 2. Load Dataset

The raw diabetes prediction dataset is loaded from the `data/raw` directory. This dataset will be cleaned and prepared for the next notebook, where machine learning models will be trained and evaluated.

```python
df = pd.read_csv("../data/raw/diabetes_prediction_dataset.csv")
```

## 3. Initial Data Check

Before applying preprocessing steps, the dataset is briefly checked again to confirm its structure, number of rows and columns, data types, duplicate records, and missing values.

This step helps ensure that the cleaning process is based on the current state of the dataset.

```python
df.info()
```

## 4. Removing Duplicate Rows

Fully duplicated rows were detected in the dataset. These rows contain exactly the same values across all features and the target variable.

Duplicate observations can introduce redundancy and may cause some patterns to be overrepresented during model training. Therefore, fully duplicated rows are removed in this step.

```python
df.duplicated().value_counts()
```

```python
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

Fully duplicated rows were removed because repeated observations can introduce redundancy and may cause some patterns to be overrepresented during model training.

```python
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

## 5. Handling Rare Categories

The `gender` column contains a very small number of observations labeled as `Other`. Since this category has very few samples, it may not provide enough information for model learning in this beginner-friendly classification project.

To simplify the preprocessing pipeline, rows with the rare `Other` gender category are removed.

```python
df["gender"].value_counts()
```

```python
df = df[df["gender"] != "Other"]
df = df.reset_index(drop=True)
```

The "Other" category in the gender column had a very small number of observations. Since it represents a rare category and may not provide enough information for model learning, it was removed to simplify preprocessing.

```python
df["gender"].value_counts()
```

## 6. Missing Value Check

Missing values are checked to determine whether any imputation or row removal strategy is required before model training.

Handling missing values is an important preprocessing step because many machine learning algorithms cannot work properly with missing data.

```python
df.isnull().sum()
```

No missing values were detected after the initial cleaning steps. Therefore, no imputation or additional row removal strategy is required for missing data.

## 8. Binary Encoding for Gender

The `gender` feature contains two categories after removing the rare `Other` category: `Female` and `Male`.

Since this feature is binary, it is converted into numerical form using binary encoding:

- Female → 0
- Male → 1

This transformation allows machine learning models to use the `gender` feature as a numerical input.

```python
encoder = LabelEncoder()

df["gender"] = encoder.fit_transform(df["gender"])
```

## 9. Feature Type Identification

Numerical and categorical features are identified to define the preprocessing strategy for the model training stage.

Numerical features can be scaled using `StandardScaler`, while categorical features such as `smoking_history` can be encoded using `OneHotEncoder`.

These transformations will be applied in the next notebook using a machine learning pipeline to avoid data leakage.

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
numeric_cols
```

```python
categorical_cols = df.select_dtypes(include=["object"]).columns
categorical_cols
```

## 10. Categorical Feature Handling Strategy

The `smoking_history` column contains multiple categories and does not have a natural numerical order. Therefore, it will not be label encoded in this notebook.

Instead, this feature will be handled using one-hot encoding in the model training notebook as part of a machine learning pipeline.

This approach prevents the model from assuming an artificial order between smoking history categories.

## 11. Save Cleaned Dataset

After completing the cleaning steps, the cleaned dataset is saved into the `data/processed` directory.

The saved dataset includes the following preprocessing steps:

- Fully duplicated rows removed
- Rare `Other` gender category removed
- Missing values checked
- `gender` feature converted into binary numerical format

The `smoking_history` feature is intentionally kept as a categorical variable and will be encoded in the model training notebook using a preprocessing pipeline.

```python
df.to_csv("../data/processed/diabetes_cleaned.csv", index=False)
```
